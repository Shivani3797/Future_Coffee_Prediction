```python
data = pd.read_csv(r"C:\Users\ADMIN\Desktop\PPt_2000_2021.csv")
data.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result=seasonal_decompose(data['PPT'], model='additionable', period=12)
result.seasonal.plot()
```

```python
result.plot()
```

```python
df = pd.DataFrame()
df['ds'] = pd.to_datetime(data['Date'])
df['y'] = data['PPT']
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=12 *10, freq='M')
```

```python
#future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'trend', 'trend_lower', 'trend_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```
